```python
import  numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=(6,3)
plt.rcParams['figure.dpi']=150
```

```python
from scipy.fftpack import fft
```

```python
n = 500
tf = 2.5
tn = np.linspace(0,tf,n)
T = tf/n
f1 = 20
f2 = 40
f3 = 80
```

```python
y = np.sin(2*np.pi*f1*tn) + np.sin(2*np.pi*f2*tn + 3) + np.sin(2*np.pi*f3*tn + 1.5) 
```

```python
yf = fft(y)
xf = np.linspace(0,1/(2*T),n//2)
```

```python
plt.figure()
plt.plot(tn,y)
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(xf, 2.0/n * np.abs(yf[:n//2]))
plt.title('FFT')
plt.xlabel('Frequênias Hz')
plt.grid(True)
plt.show()
```

```python
from random import random
```

```python
A = np.zeros(500)
for i in range(1,500):
    A[i] = y[i] + random()
```

```python
yf2 = fft(A)
xf2 = np.linspace(0,1/(2*T),n//2)
```

```python
plt.figure()
plt.plot(tn,A)
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(xf2, 2.0/n * np.abs(yf2[:n//2]))
plt.title('FFT')
plt.xlabel('Frequênias Hz')
plt.grid(True)
plt.show()
```
