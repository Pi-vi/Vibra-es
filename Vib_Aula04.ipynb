```python
import numpy as np
import matplotlib.pyplot as plt
```

Condições Iniciais:

```python
x = 0 # Deslocamento inicial
v = 5 # Velocidade inicial em m/s
```

Parâmetros:

```python
t = 0 # tempo inicial
tf = 15 # tempo final
h = 0.001 # passo de integração
n = int(tf/h) # número de pontos
mi= 0.1 # razão de amortecimento
k = 100 # rigidez
m = 1 # massa
g = 9.8 # gravidade

X=np.zeros(n)
V=np.zeros(n)
T=np.zeros(n)
F=np.zeros(n)

X[0]=x
V[0]=v
T[0]=t
```

Definição de Funções:
$$ \dot{v} = -\dfrac{k}{m} x - \mu m g $$

```python
def vp(tt,xt,vt):
    return -k/m*xt -mi*m*g*(np.sign(vt))
```

$$ \dot{x} = v $$

```python
def xp(tt,xt,vt):
    return vt
```

Integração Numérica:

```python
for i in range(1,n):
    
    k1x=xp(t,x,v)
    k1v=vp(t,x,v)
    
    k2x=xp(t+h/2,x+h*k1x/2,v+h*k1v/2)
    k2v=vp(t+h/2,x+h*k1x/2,v+h*k1v/2)
    
    k3x=xp(t+h/2,x+h*k2x/2,v+h*k2v/2)
    k3v=vp(t+h/2,x+h*k2x/2,v+h*k2v/2)
    
    k4x=xp(t+h,x+h*k3x,v+h*k3v)
    k4v=vp(t+h,x+h*k3x,v+h*k3v)
    
    t=t+h
    x=x+(h/6)*(k1x+2*k2x+2*k3x+k4x)
    v=v+(h/6)*(k1v+2*k2v+2*k3v+k4v)
    
    X[i]=x
    V[i]=v
    T[i]=t
```

Plotar Gráficos:

```python
%matplotlib inline
plt.rcParams['figure.figsize']=(6,4)
plt.rcParams['figure.dpi']=150
```

```python
plt.figure()
plt.plot(T,X)
plt.title('Massa-Mola-Coulomb')
plt.xlabel('Tempo[s]')
plt.ylabel('Amplitude[m]')
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(X,V)
plt.title('Espaço de Fase')
plt.xlabel('Amplitude[m]')
plt.ylabel('Velocidade[m/s]')
plt.grid(True)
plt.show()
```
