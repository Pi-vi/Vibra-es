```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [23,10]
plt.rcParams['font.size'] = 24
```

```python
import control as co
```

```python
G = co.tf([0.5],[1,2,15])
G
```

```python
co.pole(G)
```

## Resposta a uma Função Impulso

```python
t = np.linspace(0,10,1000)
t1,y1 = co.impulse_response(G,t)
```

```python
plt.plot(t1,y1)
plt.xlabel('tempo[s]')
plt.ylabel('Amplitude')
plt.grid(True)
```

## Resposta a um Degrau

```python
t1,y2 = co.step_response(G,t)
```

```python
plt.plot(t1,y2)
plt.xlabel('tempo[s]')
plt.ylabel('Amplitude')
plt.grid(True)
```

## Resposta a um Forçamento Qualquer

```python
t = np.linspace(0,20,1000)
w = 2
u = np.sin(w*t)
t1,y4,states = co.forced_response(G,t,u)
```

```python
plt.plot(t1,y4)
plt.xlabel('tempo[s]')
plt.ylabel('Amplitude')
plt.grid(True)
```
