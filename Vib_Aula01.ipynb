```python
from ipywidgets import interact,interactive,fixed,interact_manual
import ipywidgets as widgets
```

```python
t = np.linspace(0,10,1000)
```

```python
def f(A,w,phi):
    plt.plot(t,A*np.sin(w*t+phi))
    plt.ylim(-10,10)
    plt.grid(True)
    plt.xlabel('tempo')
    plt.ylabel('amplitude')
```

$$ f(x) = A*sin(\omega t + \phi)$$

```python
interact(f,A=(1,10),w=(1,10),phi=(0,2*np.pi))
```

```python

from scipy.fftpack import fft
```

```python
n = 500
tf = 2.5
tn = np.linspace(0,tf,n)
T = tf/n
f1 = 20
f2 = 40
f3 = 80
```

```python
y = np.sin(2*np.pi*f1*tn) + np.sin(2*np.pi*f2*tn + 3) + np.sin(2*np.pi*f3*tn + 1.5)
```

```python
yf = fft(y)
xf = np.linspace(0,1/(2*T),n//2)
```

```python
plt.figure()
plt.plot(tn,y)
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(xf, 2.0/n * np.abs(yf[:n//2]))
plt.title('FFT')
plt.xlabel('Frequênias Hz')
plt.grid(True)
plt.show()
```
